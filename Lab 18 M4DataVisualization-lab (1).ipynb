## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

QUERY = """
SELECT ConvertedComp
FROM master
"""
hist_data = pd.read_sql_query(QUERY,conn)
hist_data.plot(kind='hist')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
box_data = pd.read_sql_query(QUERY,conn)
box_data.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT * FROM master
"""
df = pd.read_sql_query(QUERY,conn)
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10,6), color='blue')
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here


QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
bub_data=pd.read_sql_query(QUERY,conn)

sns.scatterplot(data=bub_data, x='WorkWeekHrs', y='CodeRevHrs', size='Age', hue='Age', alpha=0.5, sizes=(10, 500)) 

plt.title('WorkWeekHrs and CodeRevHrs By Age') 
plt.xlabel('WorkWeekHrs') 
plt.ylabel('CodeRevHrs')

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

QUERY = """ 
SELECT DatabaseDesireNextYear, COUNT(*) as count 
from DatabaseDesireNextYear 
group by DatabaseDesireNextYear 
order by count(DatabaseDesireNextYear) DESC LIMIT 5 
""" 

df=pd.read_sql_query(QUERY,conn)  
df.set_index('DatabaseDesireNextYear', inplace=True)

colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
df['count'].plot(kind='pie', figsize=(15,6), autopct='%1.1f%%', startangle=90, shadow=True, labels= None, colors=colors_list)
plt.title('Top 5 databases that respondents wish to learn next year')
plt.axis('equal')
plt.legend(labels=df.index, loc='upper right')

plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age 
FROM master
WHERE Age between 30 and 35
"""
df= pd.read_sql_query(QUERY,conn)
stacked_data= df.groupby('Age').median()

stacked_data.plot(kind='bar', stacked=True)
plt.title('Stacked Chart of median WorkWeekHrs and CodeRevHrs for ages from 30 to 35')
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """
SELECT ConvertedComp, Age 
FROM master
WHERE Age between 45 and 60
"""
df= pd.read_sql_query(QUERY,conn)
line_data= df.groupby('Age').median()

line_data.plot(kind='line')
plt.title('The median ConvertedComp for ages from 45 to 60')
plt.ylabel('ConvertedComp')
plt.xlabel('Age')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch, COUNT(*) as MainBranch
FROM master
group by MainBranch
"""
bar_data= pd.read_sql_query(QUERY,conn)

bar_data.plot(kind='barh')
plt.title('Bar Chart of MainBranch')
plt.show()
```
